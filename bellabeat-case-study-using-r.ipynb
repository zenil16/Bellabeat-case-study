# Fixing Time and Date formats

```r
# daily activity

daily_activity$ActivityDate=as.POSIXct(daily_activity$ActivityDate, format="%m/%d/%Y %I:%M:%S %p", tz=Sys.timezone())
daily_activity$date <- format(daily_activity$ActivityDate, format = "%m/%d/%y")

# daily calories

daily_calories$ActivityDay=as.POSIXct(daily_calories$ActivityDay, format="%m/%d/%Y %I:%M:%S %p", tz=Sys.timezone())
daily_calories$date <- format(daily_calories$ActivityDay, format = "%m/%d/%y")

# daily steps

daily_steps$ActivityDay=as.POSIXct(daily_steps$ActivityDay, format="%m/%d/%Y %I:%M:%S %p", tz=Sys.timezone())
daily_steps$date <- format(daily_steps$ActivityDay, format = "%m/%d/%y")

# hourly intensities

hourly_intensities$ActivityHour=as.POSIXct(hourly_intensities$ActivityHour, format="%m/%d/%Y %I:%M:%S %p", tz=Sys.timezone())
hourly_intensities$date <- format(hourly_intensities$ActivityHour, format = "%m/%d/%y")
hourly_intensities$time <- format(hourly_intensities$ActivityHour, format = "%H:%M:%S")

# hourly steps

hourly_steps$ActivityHour=as.POSIXct(hourly_steps$ActivityHour, format="%m/%d/%Y %I:%M:%S %p", tz=Sys.timezone())
hourly_steps$date <- format(hourly_steps$ActivityHour, format = "%m/%d/%y")
hourly_steps$time <- format(hourly_steps$ActivityHour, format = "%H:%M:%S")

# sleep day

sleep_day$SleepDay=as.POSIXct(sleep_day$SleepDay, format="%m/%d/%Y %I:%M:%S %p", tz=Sys.timezone())
sleep_day$date <- format(sleep_day$SleepDay, format = "%m/%d/%y")
sleep_day$time <- format(sleep_day$SleepDay, format = "%H:%M:%S")
```

**Correlations**

We will try to find if there are any correlations between two of the following analysis.

1. Relation between Total steps and calories

2. Relation between Total steps and Total Intensity

```r
# 1. Total steps Vs Calories




# finding the correlation coefficient.

corr_coeff <- cor(daily_activity$TotalSteps, daily_activity$Calories) 



#plotting the scatterplot

ggplot(data = daily_activity, aes(x = TotalSteps, y= Calories)) + 
  geom_point(color = "blue") +
  geom_smooth(method = "lm", col = "orange") +
  theme_minimal() +
  theme(
    plot.title = element_text(face = "bold")
    ) +
  labs(title = "Relation between Total steps and calories",
       x= "Total Steps",
       y = "Calories") +
  annotate("text", x = max(daily_activity$TotalSteps), y = max(daily_activity$Calories), 
           label = paste("R =", round(corr_coeff, 2)),
           hjust = 1, vjust = -0.75)
```

**Key Findings**

* We can clearly see there is a direct correlation between total steps and the calories burnt, where R = 0.59 indicating a good amount of positive correlation.

* Also there are some outliers in the graph such as the user who had more than 30,000 steps yet he only burnt around 2,500 calories.

```r
# 2. Relation between Total steps and Total Intensity




# merging the tables hourly_intensities and hourly_steps.

intensity_step <- merge(hourly_intensities, hourly_steps, by = c("Id", "ActivityHour"))


# finding the correlation coefficient.

coeff <- cor(intensity_step$StepTotal, intensity_step$TotalIntensity)


# plotting the scatterplot.

ggplot(data = intensity_step, aes(x = StepTotal, y= TotalIntensity)) + 
  geom_point(color = "blue") +
  geom_smooth(method = "lm", col = "orange") +
  theme_minimal() +
  theme(
    plot.title = element_text(face = "bold")
  ) +
  labs(title = "Relation between Total steps and Total Intensity",
       x= "Total Steps",
       y = "Total Intensity") +
  annotate("text", x = max(intensity_step$StepTotal), y = max(intensity_step$TotalIntensity), 
           label = paste("R =", round(coeff, 2)),
           hjust = 1, vjust = -8)
```

**Key Findings**

* There is a very strong realtionship between these two as the correlation factor R = 0.9, but we need additional data to say it's a causation.

**Average Active Minutes**

We will try to plot a chart showing how much minutes has the user is spending doing the activities, whether they are very, fairly, lightly active or sedentary.

* First we will find the mean of every category to give us the average minutes.

* Grouping the data frame as per the categories and ordering them based on the average minutes spend by the users performing activities.

```r
# finding the mean

mean_Very_active <- mean(daily_activity$VeryActiveMinutes, na.rm = TRUE)
mean_Fairly_active <- mean(daily_activity$FairlyActiveMinutes, na.rm = TRUE)
mean_lightly_active <- mean(daily_activity$LightlyActiveMinutes, na.rm = TRUE)
mean_Sedentary_active <- mean(daily_activity$SedentaryMinutes, na.rm = TRUE)


#grouping the data frame

mean_df <- data.frame(
  category = c("Very Active", "Fairly Active", "Lightly Active", "Sedentary"),
  mean_value = c(mean_Very_active, mean_Fairly_active, mean_lightly_active, mean_Sedentary_active)) %>%
  group_by(category)


# ordering the data frame

mean_df$category <- factor(mean_df$category, levels = mean_df$category[order(mean_df$mean_value)])
```

Now plotting the bar chart.

```r
# plotting the bar chart

ggplot(data = mean_df, aes(x= category, y= mean_value, fill = category)) +
  geom_bar(stat = "identity") +
  labs(title = "Average Active Minutes",
       x = "Category",
       y = "Mean Value") +
  theme_minimal()
```

**Key Findings**

* As we can clearly see users spend almost 1000 minutes on average inactive which is very unhealthy.

* Most users spend time doing light activities rather than high energy activities.

* Another trend we can notice is the users are spending more minutes on very active activities than fairly active.

**Comparing Asleep time and time spent in bed**

* First we will be summarizing the sleep_day table and create a new data frame called sleep_summarized which contains total asleep time and total time in bed grouped by their individual Id.

```r
# summarizing

sleep_summarized <- sleep_day %>%
  group_by(Id) %>%
  summarise(Asleep_time = sum(TotalMinutesAsleep, na.rm = TRUE), Bed_time = sum(TotalTimeInBed, na.rm = TRUE))


# printing the data frame

print(sleep_summarized)
```

* We will be summarizing the intensity_step table as we need the total steps fromn that table.

```r
# summarizing

step_summarized <- intensity_step %>%
  group_by(Id) %>%
  summarise(step_sum = sum(StepTotal, na.rm = TRUE))

# printing the data frame

print(step_summarized)
```

* Merging both the data frames

```r
#Merging and printing

merged_sleep_step <- merge(sleep_summarized, step_summarized, by = "Id")

print(merged_sleep_step)
```

We can see we have 24 unique entries with their respective asleep time, time in bed and total steps.

* transforming the data frame from wide format to long format.

```r
# transforming

df_long <- merged_sleep_step %>%
  pivot_longer(cols = c(Asleep_time, Bed_time),
               names_to = "metric",
               values_to = "value")

#printing

print(df_long)
```

* Plotting the group bar comparing asleep time and time in bed.

```r
# plotting

ggplot(df_long, aes(x = reorder(factor(Id), value), y = value, fill = metric)) + 
  geom_bar(stat = "identity", position = "dodge") +
  labs(title = "Comparing Asleep Time and Time in Bed",
       x = "User Id", 
       y = "Total Minutes",
       fill = "Metric") + 
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust = 1)) 
```

**Key findings**

* We can clearly see many of the users are facing difficulties to sleep, if we put the light on a particular user Id, out of all most 208 hours in the bed they only slept for approximately 127 hours, which means almost 40% of the time they were awake.

**Average Total intensity VS Time**

Let's see at what time are the users most active.

```r
# creating a new data frame grouped by time

new_int <- hourly_intensities %>%
  group_by(time) %>%
  drop_na() %>%
  summarise(mean_total_int = mean(TotalIntensity))

# Plotting the histogram

ggplot(data = new_int, aes(x = time, y = mean_total_int)) +
  geom_histogram(stat = "identity", fill = "darkgreen") + 
  theme(axis.text.x = element_text(angle = 90)) +
  labs(title = "Average Total Intensity Vs Time")
```

**Key Findings**

* We can see that most of the users were active from 7 am to 9 pm.
